```python
import pandas as pd
data = {  "Month" : ['Jan', 'Feb','Mar', 'Apr', 'May','Jun'],
        "Sales" : [10000, 12000, 15000, 13000, 17000, 16000 ],
        "Profit" : [2000,3000,4000,2500,3500,3000]}
df = pd.DataFrame(data)
print(df)
```

#1 LINE PLOT MONTHLY SALES


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], color='blue',marker='o',linestyle='-' ,label='Sales')
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,3))
plt.plot(df['Month'], df['Sales'], color='blue',marker='o',linestyle='-' ,label='Sales')
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.legend()
plt.show()
```

#2 BARPLOT MONTH VS PROFIT

```python
plt.figure(figsize=(6,3))
plt.bar(df['Month'], df['Sales'], color='skyblue' ,label='Sales')
plt.bar(df['Month'], df['Profit'],color='green', label='profit', bottom=df['Sales'])
plt.title('Sales Trends by Months')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#3 PIECHART PROFIT VS MONTH

```python
from enum import auto
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.pie(df['Profit'] ,labels=df['Month'], autopct= ('%1.2f%%'), startangle=140, colors= plt.cm.Paired.colors)
plt.title('Profit by Month')
plt.show()
```

# 4 SCATTER PLOT

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Sales'], df['Profit'], color ='green', s =100, edgecolors='black')
plt.title('Sales vs Profit Scatter Plot')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.grid(True)
plt.show()

```

# Histogram

```python
plt.figure(figsize=(8,4))
plt.hist(df['Sales'], bins=5, color ='purple', edgecolor='black'),
plt.title('Sales Distribution'),
plt.xlabel('Sales'),
plt.ylabel('Frequency'),
plt.tight_layout(),
plt.show()
```

# BOX PLOT


```python
plt.figure(figsize=(8,4))
plt.boxplot(df['Profit']),
plt.title('Sales Box Plot'),
plt.ylabel('Sales'),
plt.tight_layout(),
plt.show()

```
